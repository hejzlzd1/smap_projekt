```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.utils import image_dataset_from_directory as datasetImport
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# Define constants and paths
data_dir = "SmallDataset"
data_dir_grayscale = "SmallDatasetGrayscale"
batch_size = 32
image_size = (300, 300)
num_classes = 7
epochs = 50
random_seed = 42
features = 64

def load_dataset(color_mode='rgb', subset='training', dir='SmallDataset'):
    return datasetImport(
        dir,
        color_mode=color_mode,
        seed=random_seed,
        image_size=image_size,
        batch_size=batch_size,
        subset=subset,
        validation_split=0.2,
    )

train_dataset_rgb = load_dataset(color_mode='rgb', dir=data_dir)
test_dataset_rgb = load_dataset(color_mode='rgb', subset='validation', dir=data_dir)
train_dataset_grayscale = load_dataset(color_mode='rgb', dir=data_dir_grayscale)
test_dataset_grayscale = load_dataset(color_mode='rgb', subset='validation', dir=data_dir_grayscale)

def build_model(input_shape, num_classes):
    data_augmentation = tf.keras.Sequential([
      keras.layers.RandomFlip("horizontal_and_vertical"),
      keras.layers.RandomRotation(0.15),
      keras.layers.RandomContrast(0.05),
    ])
    base_model = keras.applications.EfficientNetV2B3(weights='imagenet', include_top=False, input_shape = input_shape)
    base_model.trainable = True
    fine_tune_at = 70
    for layer in base_model.layers[:fine_tune_at]:
        layer.trainable = False
    
    model = keras.Sequential([
        data_augmentation,
        base_model,
        keras.layers.Dropout(0.25),
        keras.layers.GlobalAveragePooling2D(),
        keras.layers.BatchNormalization(),
    ])

    # Add dense layers
    for units in [features * 4, features * 2, features]:
        model.add(keras.layers.Dense(units, kernel_initializer='he_uniform', kernel_regularizer=keras.regularizers.l2(0.002)))
        model.add(keras.layers.Activation('relu'))
        model.add(keras.layers.Dropout(0.25))
        model.add(keras.layers.BatchNormalization())

    model.add(keras.layers.Dense(num_classes, activation='softmax'))
    return model

model_rgb = build_model(image_size + (3,), num_classes)
model_grayscale = build_model(image_size + (3,), num_classes)

lrd = keras.callbacks.ReduceLROnPlateau(monitor = 'val_loss',patience = 15,verbose = 1,factor = 0.50, min_lr = 1e-6)
es = keras.callbacks.EarlyStopping(verbose=1, patience=15)

def train_model(model, train_dataset, test_dataset, name):
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=2e-3),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    print(f'\n----------- \n Training {name} model:')
    history = model.fit(train_dataset, validation_data=test_dataset, epochs=epochs, callbacks=[lrd, es])

    # Plot accuracy and loss
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    # Evaluate the model on test data
    test_loss, test_accuracy = model.evaluate(test_dataset,)
    print(f'Validation Accuracy ({name}): {test_accuracy:.2f}')

    # Calculate and display a confusion matrix and classification report
    y_true, y_pred = [], []
    for images, labels in test_dataset:
        y_true.extend(labels.numpy())
        predictions = model.predict(images, verbose=0)
        y_pred.extend(np.argmax(predictions, axis=1))

    cm = confusion_matrix(y_true, y_pred)
    
    print(f'Confusion Matrix ({name} model):')
    cm_display = ConfusionMatrixDisplay(confusion_matrix = cm)
    plt.figure(figsize = (10,10))
    cm_display.plot(cmap = 'viridis', values_format = 'd')

    plt.title('Confusion Matrix', fontsize = 20)
    plt.show()

    
    cr = classification_report(y_true, y_pred, target_names=train_dataset.class_names)

    print(f'Classification Report ({name} model):')
    print(cr)

    # Save the model 
    model.save(f'{name}_model_efficientnetV2B3.h5')
    print(f'{name} model saved.')

    return test_loss, test_accuracy, cm, cr

test_loss_rgb, test_accuracy_rgb, cm_rgb, cr_rgb = train_model(model_rgb, train_dataset_rgb, test_dataset_rgb, 'RGB')
test_loss_grayscale, test_accuracy_grayscale, cm_grayscale, cr_grayscale = train_model(model_grayscale, train_dataset_grayscale, test_dataset_grayscale, 'Grayscale')
```
